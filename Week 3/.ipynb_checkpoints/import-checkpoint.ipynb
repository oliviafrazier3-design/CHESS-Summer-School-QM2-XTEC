```python
import os
import glob
import h5py

base_path = "/nfs/chess/id4baux/2025-2/he-4025-d/nxrefine/Ta2NiSeS70/S1"
subdirs = sorted([d for d in os.listdir(base_path) if d.isdigit()], key=int)

# Storage dictionary for memory-mapped file pointer handles
lazy_datasets = {}
DATASET_PATH = "entry/data/n"

print("--- Starting Instant Lazy Data Import ---")
for folder in subdirs:
    folder_path = os.path.join(base_path, folder)
    nxs_files = glob.glob(os.path.join(folder_path, "*.nxs"))
    
    if nxs_files:
        file_path = nxs_files[0]
        try:
            # Open file handle and leave it open
            f = h5py.File(file_path, 'r')
            if DATASET_PATH in f:
                # CRITICAL: No [:] at the end. This references the data on disk without reading it.
                lazy_datasets[folder] = f[DATASET_PATH]
                print(f"Mapped Folder {folder:3s} | Shape: {lazy_datasets[folder].shape} (0 MB RAM used)")
        except Exception as e:
            print(f"Failed folder {folder}: {e}")

print(f"\nImport finished! Loaded {len(lazy_datasets)} pointers securely.")
```

```python
import dask.array as da

# 1. Sort the subdirectories numerically so your stacked axis matches ascending temperature
sorted_folders = sorted(lazy_datasets.keys(), key=int)

# 2. Convert each h5py dataset handle into a lazy Dask array pointer
lazy_arrays = [da.from_array(lazy_datasets[folder], chunks=lazy_datasets[folder].chunks) 
               for folder in sorted_folders]

# 3. Stack them cleanly along a brand new axis (Axis 0 = Temperature)
master_stacked_array = da.stack(lazy_arrays, axis=0)

print("--- Virtual Matrix Stacked Successfully ---")
print("Combined 4D Shape:", master_stacked_array.shape)
print("Data Type:", master_stacked_array.dtype)
print("RAM Used: 0 MB (Array remains safely on disk until called!)")
```
